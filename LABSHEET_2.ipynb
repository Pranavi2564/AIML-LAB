```python
import matplotlib.pyplot as plt
```

###1. LINE CHART

```python
yield_Apples=[0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(yield_Apples)
```

```python
years=[2010,2011,2012,2013,2014,2015]
yield_Apples=[0.895,0.91,0.919,0.926,0.929,0.931]

plt.plot(years,yield_Apples);
```

```python
plt.plot(years,yield_Apples)
plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')
plt.title('YIELD vs YEAR')
```

PLOTTTING MULTIPLE LINES

```python
year=range(2000,2008)

apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]
```

```python
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield (tons per hectare)');
```

CHART TITLE AND LEGEND

```python
year=range(2000,2008)

apples=[0.895,0.91,0.919,0.926,0.929,0.934,0.936,0.937]
oranges=[0.962,0.941,0.930,0.918,0.908,0.907,0.904,0.901]

plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')
plt.title('Crop yields in Kanto')
plt.legend(['Apples','Oranges']);
```

LINE MARKERS

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')

plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')
plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

STYLING LINES AND MARKERS

```python
plt.plot(year,apples,marker='s',c='b',ls='-',lw=2,ms=8,mew=2,mec='navy',alpha=0.5)
plt.plot(year,oranges,marker='o',c='r',ls='--',lw=3,ms=10,alpha=0.5,mew=2,mec='red')

plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
#fmt='[marker][line][color]'
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')

plt.xlabel('year')
plt.ylabel('yield (tons per hectare)')

plt.title('Crop Yields in Kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,oranges,'or')
plt.title('Yield of Oranges');
```

CHANGING THE FIGURE SIZE

```python
plt.figure(figsize=(10,5))

plt.plot(year,oranges,'or')
plt.title('Yield of Oranges');
```

###2. BAR CHART

```python
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]

plt.bar(x,y)
plt.title("BAR CHART")
plt.legend(["Bar"])
plt.show()
```

###3. SCATTER CHART

```python
x = [3,1,3,12,2,4,4]
y = [3,2,1,4,5,6,7]

plt.scatter(x,y)

plt.legend("A")
plt.xlabel('x')
plt.ylabel('y')

plt.title("SCATTER CHART");
```

###4. PIE CHART

```python
x = [1,2,3,4]

e = (0.1,0,0.3,0)

plt.pie(x, explode = e)

plt.title("PIE CHART")
plt.show()
```
